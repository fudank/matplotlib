```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(0,1,100)
x = z*np.sin(z*25)
y = z*np.cos(z*25)
ax.plot3D(x,y,z,'gray')
ax.set_title('3d line plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(0,1,10)
x = z*np.log10(z*25)
y = z*np.log10(z*25)
ax.plot3D(x,y,z,'gray')
ax.set_title('3d line plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(0,1,100)
x = z*np.sin(z*25)
y = z*np.cos(z*25)
c = x+y
ax.scatter(x,y,z,'r',c = c)
ax.set_title('3d line plot')
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('z_axis')
plt.show()
```

```python
def f(x1,y1):
    return np.sin(np.sqrt(x1**2 + y1**2))
x1 = np.linspace(-6,6,40)
y1 = np.linspace(-6,6,40)

x1,y1 = np.meshgrid(x1,y1)
z1 = f(x1,y1)

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.contour3D(x1,y1,z1,40,cmap = 'binary')
ax.set_title('3d line plot')
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('z_axis')
plt.show()

```

```python
def f(x1,y1):
    return np.sin(np.sqrt(x1**2 + y1**2))
x1 = np.linspace(-6,6,40)
y1 = np.linspace(-6,6,40)

x1,y1 = np.meshgrid(x1,y1)
z1 = f(x1,y1)

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_wireframe(x1,y1,z1,color = 'b')
ax.set_title('3d line plot')
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('z_axis')
plt.show()
```

```python
def f(x1,y1):
    return np.sin(np.sqrt(x1**2 + y1**2))
x1 = np.linspace(-6,6,40)
y1 = np.linspace(-6,6,40)

x1,y1 = np.meshgrid(x1,y1)
z1 = f(x1,y1)

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_surface(x1,y1,z1,color = 'r', cmap = 'viridis',edgecolor = 'r')
ax.set_title('surface')
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('z_axis')
plt.show()
```
