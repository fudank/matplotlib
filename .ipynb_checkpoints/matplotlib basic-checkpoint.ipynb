```python
import numpy as np
fig ,a  = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
import numpy as np
fig ,a  = plt.subplots(3,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[1][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[0][1].plot(x,np.log10(x))
a[1][1].set_title('log')
a[2][0].plot(x,sin(x))
a[2][0].set_title('sine function')
a[2][1].plot(x,tan(x))
a[2][1].set_title('tan function')
plt.show()
```
